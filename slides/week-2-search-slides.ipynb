## Informed searches

* Think: How do we order the priority queue to get the behavior we want

* Priority queues are ordered by numeric values

* Values come from the _function(node)_

* Function evaluates the node and returns a value

* Return value is based on the state the node represents

* Examples
    - Cost of path to current state
    - Heuristic estimate additional cost to reach the goal
    - Combination: Cost + Estimate

# Informed search algorithms

Queue ordering has a big effect as we just saw.  How better order our Queue to find solutions efficiently?

1. Best First Search -- Uses a priority queue ordered by a function

2. _Uniform Cost Search_ -- Uses least-cost path -- g(n)

3. Greedy search -- Uses least-cost heuristic estimate to goal -- h(n)

4. _A* Search_ -- Uses the sum of the least-cost path and heuristic -- g(n) + h(n)

## Informed search algorithms

* Uniform cost search == BFS + g(lowest cost path)
    - g() == cost incurred so far
    - Finds the optimal path

* Greedy search == BFS + h(estimated remaining cost)
    - h() == heuristic measure that must be greater than or equal to actual cost
    - Not optimal but it is efficient

* A* == BFS + g(accumulated cost) + h(estimated remaining cost)
    - Finds the optimal path
    - Complete == If a solution exists, it will find it
    - Efficient
    - Heuristics are problem specific

## Solution notes

* Problem: Get from Sibiu to Bucharest. 
* Successors of Sibiu are Rimnicu Vilcea and Fagaras, with costs 80 and 99. 
* Least-cost node, Rimnicu Vilcea, is expanded
* next, adding Pitesti with cost 80 + 97 = 177. 
* Now least-cost node is Fagaras, so it is expanded
* Add Bucharest with cost 99 + 211 = 310. 
* Now a goal node has been generated,
* but uniform-cost search keeps going
* Choosing Pitesti for expansion and adding a second path to Bucharest
* Now cost to Bucharest is 80 + 97 + 101 = 278 
* Algorithm checks to see if this new path is better; it is, so the old one is discarded. 
* Bucharest, now with g-cost 278 is selected for expansion and the solution is returned.

# Heuristics

## Admissible Heuristic Definition

* Admissible:  H(x, g) <= P(x, g) where H = heuristic, P = path cost

* As the "bird flies" distance is admissible

* It always less than or equal to a road distance

* Only works for geographically search problems like Google Maps


## Consistent Heuristic

| H(x,g) - H(y, g) | <= P(x,y)
