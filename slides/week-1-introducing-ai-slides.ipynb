## Early AI History

* During the industrial revolution, there is great interest in creating machines to behave like humans or animals. [Cite: Frolicsome engines the long prehistory of artificial intelligence](https://publicdomainreview.org/2016/05/04/frolicsome-engines-the-long-prehistory-of-artificial-intelligence/).

* [Ada Lovelace](https://www.wired.com/2015/12/untangling-the-tale-of-ada-lovelace/) creates the programmings for mechanical calculating devices -- including her well documented [approach to calculating Bernoulli numbers](https://rclab.de/rclab/_media/analyticalengine/aal_noteg_glaschick_v1.2.pdf) -- in her partnership with Charles Babbage.  
* Ada said, "The Analytical Engine weaves algebraic patterns, just as the Jacquard loom weaves flowers and leaves." [Ada Lovelace: Weaving Algebraic Patterns ](http://themorrisian.blogspot.com/2012/12/ada-lovelace-weaving-algebraic-patterns.html)

* The industrial revolution lasts from roughly 1760 to 1820.  Automating cloth production using the new Jacquard loom late in this period displaces workers who begin taking out their frustrations by sabotaging equipment. See a brief history of the [Luddite riots](http://www.nationalarchives.gov.uk/education/politics/g3/).

* Not much happened until WWII made faster calculations a necessity.  Much early work occurs at the [Army Research Laboratory (computing history link)](https://www.arl.army.mil/www/default.cfm?page=148), including [ENIAC](http://www.computerhistory.org/revolution/birth-of-the-computer/4/78).

* Initially, single-purpose computers were used to calculate things like firing tables for artillery.  Alan Turing used specialized machine -- [the British Bombe](http://www.cryptomuseum.com/crypto/bombe/) -- to help break the German Enigma code. Estimates suggest his efforts shortened the war by several years.

* Turing's in 1937 had proposed a "logical computing machine" to prove that some values are not computable. This idea becomes known as a Turing machine, and it starts the field of theoretical computer science. _Cite: Isaacson, W. (2015). The Innovators: How a group of hackers, geniuses, and geeks created the digital revolution. New York: Simon & Schuster Paperbacks. p44-45._

* Later during WWII, [John Von Neumann](http://www.computinghistory.org.uk/det/3665/john-von-neumann/) uses the Turning machine as inspiration in developing ENIAC -- the first that stores code and data in memory.  The Von Neumann architecture is _the_ architecture used in all computers since 1945.  Only now with quantum computers are new architectures emerging.  _Cite: Isaacson, W. (2015). The Innovators: How a group of hackers, geniuses, and geeks created the digital revolution. New York: Simon & Schuster Paperbacks. p45-47._

* In 1950, Alan Turing proposes the Turning Test -- a practical test of thinking abilities of a machine.  His paper starts what will become known as Artificial Intelligence.  But by that time Turning has committed suicide two years after having been prosecuted for being gay.  In 2013, he is pardoned by the Queen. _Cite: Isaacson, W. (2015). The Innovators: How a group of hackers, geniuses, and geeks created the digital revolution. New York: Simon & Schuster Paperbacks. p122-129._   

## Conceptualizing AI

#### Thinking humanly -- Cognitive science
* Need to understand how humans think first
    - Introspection
    - Psychological experiments
    - Observe the brain in action 
* Cognitive science uses AI models of the brain and psychological experiments to construct testable theories
* Conclusion: testing the human  brain is difficult

#### Acting humanly -- The Turing's Test
* Operational definition proposed by Alan Turing in 1950
* Idea: Consider a machine to be intelligent, if can fool a human into thinking it is human

```python
Image('images/Turing_test_diagram.png')
```

* Can C tell whether A or B is the machine? No = machine is intelligent, or Yes = probably not

#### Thinking rationally -- A reasoned approach
* Make Boolean statements about the world / environment
* True or false or unprovable
* Represent the state of the world by building sets of assertions
* Do live in a provable world?  

#### Acting rationally -- Intelligent agent approach
* Acting to achieve a goal using what we know about the world
* Able to deal with uncertainty when logical reasoning fails
* Hypothesis testing is well supported
* Modern AI resides here!

## Relating rationality and performance?

* Performance measure defines the cost of achieving the goal

* Allows different approaches to be compared

* What effects performance?

    - Agent's prior knowledge of the environment

    - Actions available to the agent and their cost

    - Agent's precept sequence and what the agent does with it

* Rational agents attempt to maximize their performance by taking the fewest actions to achieve the goal.

## Task environment for a self-driving car

In order to reach the goal, agents need to be designed for the function in there environment.

Performance, Environment, Actuators, Sensors (PEAS) is good way to specify the requirements the agent needs to fulfill.

*   **Performance**: Safe, legal, comfortable, maximize profits
*   **Environment**: Roads, traffic pedestrians, customers
*   **Actuators**: Steering, accelerate, brakes, signal, horn
*   **Sensors**: Cameras, sonar, speed, direction, motor, ...

## Qualities of an environment

* Fully or partially observable --> Is entire state of the world knowable?

* Deterministic vs Stochastic --> Is the world uncertain?

* Episodic vs sequential --> Do previous actions matter?

* Static vs dynamic --> Do changes occur while picking action?

* Discrete vs continuous --> Finite or infinite possibilities?

* Known vs unknown --> Are the rules known? 

## Additional qualities

* Autonomy --> The extent that agents learn from the environment vs pre-programming

* Omniscience --> Knowing the answer is advance.  Data leakage  in the context of AI

## State Space - All states possible in an Environment

* Example: Chess

    - 16 pieces, 8*8 board

    - About 30 legal moves per player per turn

    - 30*30 = 900 possible moves per round

    - Round 900 up to 1000 for ease of calculation

    - Average game lasts 40 rounds

    - 10^3*40 = 10^120
